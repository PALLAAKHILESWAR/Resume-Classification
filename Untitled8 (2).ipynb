```python
import os
import pandas as pd
from docx import Document
import fitz  # for PDF

def read_docx(file_path):
    """Reads text from a .docx file."""
    try:
        doc = Document(file_path)
        return '\n'.join([para.text for para in doc.paragraphs])
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return None

def read_pdf(file_path):
    """Reads text from a .pdf file."""
    text = ""
    try:
        doc = fitz.open(file_path)
        for page in doc:
            text += page.get_text()
        return text
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return None

root_path = "/content/drive/MyDrive/Resumes_Docx"
data = []

for folder in os.listdir(root_path):
    folder_path = os.path.join(root_path, folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            content = None  # Initialize content

            if file.endswith(".docx"):
                content = read_docx(file_path)
            elif file.endswith(".pdf"):
                content = read_pdf(file_path)
            # Add handling for .doc files
            elif file.endswith(".doc"):
                 print(f"Skipping .doc file: {file_path} (unsupported format by python-docx)")
                 continue # Skip unsupported formats for now
            else:
                print(f"Skipping file with unsupported format: {file_path}")
                continue  # Skip other unsupported formats

            if content is not None:
                data.append({
                    "file_name": file,
                    "resume_content": content,
                    "folder_name": folder
                })

df = pd.DataFrame(data)
print(df.head())
```

```python
import re
import string

def clean_test(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)  # remove extra spaces
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub('\[.*?\]', ' ', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\w*\d\w*', ' ', text)
    text = re.sub('[0-9' ']+'," ",text)
    text = re.sub('[‘’“”…]', ' ', text)
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'http\S+|www\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\t', ' ', text)
    text = re.sub('\r', ' ', text)
    text = re.sub(' +', ' ', text)
    text = re.sub('^ +', '', text)
    text = re.sub(' +$', '', text)
    text = re.sub('^ +| +$', '', text)
    text = text.strip()
    return text

clean = lambda x: clean_test(x)
```

```python
#Bi-gram
def get_top_n2_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(2,2),  #for tri-gram, put ngram_range=(3,3)
            max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1],
                reverse=True)
    return words_freq[:n]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
top2_words = get_top_n2_words(df['resume_content'], n=20)
top2_df = pd.DataFrame(top2_words, columns=['Bi-gram', 'Freq'])
top2_df.head()
```

```python
#Bi-gram plot
import matplotlib.pyplot as plt
import seaborn as sns
top20_bigram = top2_df.iloc[0:20,:]
fig = plt.figure(figsize = (10, 5))
plot=sns.barplot(x=top20_bigram["Bi-gram"],y=top20_bigram["Freq"])
plot.set_xticklabels(rotation=45,labels = top20_bigram["Bi-gram"])
```

```python
#Tri-gram
def get_top_n3_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(3,3),
           max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1],
                reverse=True)
    return words_freq[:n]
```

```python
top3_words = get_top_n3_words(df['resume_content'], n=200)
top3_df = pd.DataFrame(top3_words)
top3_df.columns=["Tri-gram", "Freq"]
```
